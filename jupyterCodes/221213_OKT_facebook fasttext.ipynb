```python
#epoch : 전체 데이터를 몇 번 사용해서 학습할껀지 
#thread 수 : 14개 이하로 설정
model = ft.train_unsupervised('train.txt', model='skipgram',minn=2, maxn=4, min_count=50, epoch=10, wordNgrams=2, thread=10)
```

```python
model.save_model('room_model.bin')
```

```python
vectors = []
for i in data.doc_new:
    vectors.append(model.get_word_vector(i))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 비슷한 매물 상위 10개 추천
input_data = input('찾고자 하는 매물의 정보를 입력해주세요.:')
input_data = input_data.replace(" ", "")

#print(input_data)

input_vector = model.get_word_vector(input_data)
cosine_similarities = cosine_similarity(vectors, [input_vector])
sim_scores = list(enumerate(cosine_similarities))
sim_scores = sorted(sim_scores, key = lambda x : x[1], reverse = True)

rec_index = [i[0] for i in sim_scores]
rec_index = rec_index[:10]

recommend = data.iloc[rec_index]
recommend
```
