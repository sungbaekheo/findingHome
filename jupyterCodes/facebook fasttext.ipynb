```python
import pandas as pd
import numpy as np
import fasttext as ft
```

```python
data = pd.read_pickle('seoul_oneroom_for_ml.pkl')
```

```python
data.isnull().sum()
```

```python
text_file = open(r'train.txt', 'w', encoding='utf-8-sig')
my_list = "\n".join(data.doc)
text_file.write(my_list)
text_file.close()
```

```python
result = data.token
```

```python
result
```

```python
model = ft.train_unsupervised('train.txt', model='skipgram',minn=2, maxn=3, min_count=5, epoch=10, wordNgrams=2, thread=12)
```

```python
model.words
```

```python
model.save_model('room_model.bin')
```

```python
model.get_word_vector('king')
```

```python
vectors = []
for i in data.doc:
    vectors.append(model.get_word_vector(i))
```

```python
len(vectors)
```

```python
data['vector'] = vectors
```

```python
data
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(['index'], axis=1, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 비슷한 매물 상위 20개 추천
input_data = input('찾고자 하는 매물의 정보를 입력해주세요.:')

input_vector = model.get_word_vector(input_data)
cosine_similarities = cosine_similarity(vectors, [input_vector])
sim_scores = list(enumerate(cosine_similarities))
sim_scores = sorted(sim_scores, key = lambda x : x[1], reverse = True)

rec_index = [i[0] for i in sim_scores]
rec_index = rec_index[:20]

recommend = data.iloc[rec_index]
recommend
```

```python
for i in recommend['doc']:
    print(i)
```
