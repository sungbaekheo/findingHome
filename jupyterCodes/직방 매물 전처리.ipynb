```python
import pandas as pd
import numpy as np
```

```python
house_info = pd.read_pickle('seoul_room_renewed.pkl')
```

```python
house_info
```

```python
house_info['info'][4]['설명']['요약'] + " " + house_info['info'][4]['설명']['상세내용']
```

```python
for i in house_info['info']:
    if i['제목'] is None:
        i['제목'] = 'None'
```

```python
urls = []
address = []
details = [] # 제목 + 요약 + 상세내용
pics = []
types = []
deposits = []
rental_fees = []
maintenance_fees = []

for i in house_info['info']:
    urls.append(i['url'])
    address.append(list(i['주소']))
    details.append(i['제목'] + " " + i['설명']['상세내용'])
    pics.append(i['정보']['사진'][0])
    types.append(i['정보']['전세/월세'])
    deposits.append(i['비용']['보증금액'])
    rental_fees.append(i['비용']['월세금액'])
    maintenance_fees.append(i['비용']['관리비'])
    
```

```python
print(len(urls))
print(len(details))
print(len(pics))
print(len(types))
print(len(deposits))
print(len(rental_fees))
print(len(maintenance_fees))
```

```python
for i in range(len(address)):
    address[i].remove(None)
    address[i][0] = address[i][0].replace('서울특별시', '서울시')
    address[i] = " ".join(address[i])
```

```python
address
```

```python
for i in address:
    i[0].replace('서울특별시', '서울시')
```

```python
address
```

```python
house_info['url'] = urls
house_info['address'] = address
house_info['detail'] = details
house_info['picture'] = pics
house_info['type'] = types
house_info['deposit'] = deposits
house_info['rental_fee'] = rental_fees
house_info['maintenance_fee'] = maintenance_fees
```

```python
house_info = house_info[house_info['address'].str.contains('서울')]
```

```python
house_info.to_csv('seoul_oneroom_renew.csv', encoding = 'utf-8-sig')
house_info.to_pickle('seoul_oneroom_renew.pkl')
```
